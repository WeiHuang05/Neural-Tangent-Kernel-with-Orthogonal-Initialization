# Figures 3

Generate figures of dynamics of full batch gradient descent on Gaussian and orthogonal initialized networks.

## Setup

```python
from __future__ import print_function
import os, sys
from tqdm import tqdm
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
```

## $N = 400$ and $L=3$ Orthogonal network

```python
import csv   

number_av = 30; number_st = 10000
Ort_N400_train_acc = np.zeros((number_av,number_st)); Ort_N400_train_los = np.zeros((number_av,number_st))
Ort_N400_test_acc = np.zeros((number_av,number_st)); Ort_N400_test_los = np.zeros((number_av,number_st))

for i in range(number_av):
    j = 0

    with open('../data/full_batch/tanh_orth_L3_N400/tanh_orthogonal_L3_N400_lr_%d.csv' %i , newline='') as csvfile:
        spamreader = csv.reader(csvfile, delimiter=' ')
        for row in spamreader:
            
            Ort_N400_train_acc[i,j] = row[2]; Ort_N400_train_los[i,j] = row[3]
            Ort_N400_test_acc[i,j] = row[4];  Ort_N400_test_los[i,j] = row[5] 
            j = j+1
    
Ort_N400_train_acc_av = np.mean(Ort_N400_train_acc[:,:],0)
Ort_N400_test_acc_av = np.mean(Ort_N400_test_acc[:,:],0)

Ort_N400_train_los_av = np.mean(Ort_N400_train_los[:,:],0)
Ort_N400_test_los_av = np.mean(Ort_N400_test_los[:,:],0)
```

# Figure 3:  Dynamics of full batch gradient descent on dataset
Dynamics of full batch gradient descent on Gaussian and orthogonal initialized networks of $T=10^4$ steps. Orthogonal networks behaves similarly to dynamics on the corresponding Gaussian networks, for loss and accuracy functions. The dataset is selected from full CIFAR10 with $D=256$, while MSE loss and tanh fully-connected networks are adopted for the classification task. (a)(b) Network with length $L=3$ and width of $n = 400$, with $\sigma^2_w = 1.5$, and $\sigma^2_b =0.01$. (c)(d) Network with length $L=7$ and width of $n = 800$, with $\sigma^2_w = 1.5$, and $\sigma^2_b =0.1$. While the solid lines stand for Gaussian weights, dotted lines represent orthogonal initialization.

```python

plt.figure(figsize=(15, 3))
gs = plt.GridSpec(1, 4)
matplotlib.rc('xtick',labelsize=15)
matplotlib.rc('ytick',labelsize=15)
import seaborn as sns
sns.set_style("darkgrid", {"axes.facecolor":".95"})

a1 = plt.subplot(gs[0, 0])

plt.loglog(Gau_N400_train_acc_av_l1,color='blue',label = 'Train Gaussian')
plt.loglog(Ort_N400_train_acc_av_l1,color='blue',ls='--',label ='Train Orthogonal')

plt.semilogx(Gau_N400_test_acc_av_l1,color='red',label ='Test Gaussian')
plt.semilogx(Ort_N400_test_acc_av_l1,color='red',ls='--',label ='Test Orthogonal')

plt.ylim(5e-2, 2e0); plt.legend(fontsize=10); 
plt.xlabel('$t$',fontsize=15); 
plt.title('(a) L=3, Accuracy',fontsize=15)



a2 = plt.subplot(gs[0, 1])

plt.semilogx(Gau_N400_train_los_av_l1,color='blue')
plt.semilogx(Ort_N400_train_los_av_l1,color='blue',ls='--')

plt.loglog(Gau_N400_test_los_av_l1,color='red')
plt.loglog(Ort_N400_test_los_av_l1,color='red',ls='--')

plt.ylim(1e-1, 1e1)
plt.xlabel('$t$',fontsize=15); 
plt.title('(b) L=3, Loss',fontsize=15)

a3 = plt.subplot(gs[0, 2])

plt.loglog(Gau_N800_train_acc_av,color='blue')
plt.loglog(Ort_N800_train_acc_av,color='blue',ls='--')

plt.semilogx(Gau_N800_test_acc_av,color='red')
plt.semilogx(Ort_N800_test_acc_av,color='red',ls='--')
plt.ylim(5e-2, 2e0)
plt.xlabel('$t$',fontsize=15); 
plt.title('(c) L=7, Accuracy',fontsize=15)

a4 = plt.subplot(gs[0, 3])

plt.semilogx(Gau_N800_train_los_av,color='blue')
plt.semilogx(Ort_N800_train_los_av,color='blue',ls='--')

plt.loglog(Gau_N800_test_los_av,color='red')
plt.loglog(Ort_N800_test_los_av,color='red',ls='--')

plt.ylim(1e-1, 1e1)
plt.xlabel('$t$',fontsize=15); 
plt.title('(d) L=7, Loss',fontsize=15)

plt.savefig("fig3_smallnet.pdf", bbox_inches='tight')
```
