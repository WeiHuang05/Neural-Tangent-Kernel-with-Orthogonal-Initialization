# Figure 2:  Changes of weights and empirical NTK
Changes of weights, empirical NTK on a three hidden layer Erf Network. Solid lines correspond to empirical simulation and dotted lines are theoretical predictions, i.e. black dotted lines are $1/\sqrt{n}$ while red dotted lines are $1/n$. (a) weight changes on Gaussian initialized network. (b) weight changes on the orthogonal initialized network. (c) NTK changes on both Gaussian and orthogonal networks.

```python
import matplotlib
plt.figure(figsize=(20,5))
gs = plt.GridSpec(1,3)
matplotlib.rc('xtick',labelsize=15)
matplotlib.rc('ytick',labelsize=15)
x = np.arange(number_width)
import seaborn as sns
sns.set_style("darkgrid", {"axes.facecolor":".95"})

ax0 = plt.subplot((gs[0,0]))
plt.errorbar(2**(x+4),np.mean(gau_w1,1),np.var(gau_w1,1),lw=2,color='blue', alpha=0.5,label='layer 1')
plt.errorbar(2**(x+4),np.mean(gau_w2,1),np.var(gau_w2,1),lw=2,color='brown',  alpha=0.5,label='layer 2')
plt.errorbar(2**(x+4),np.mean(gau_w3,1),np.var(gau_w3,1),lw=2,color='green', alpha=0.5, label='layer 3')
plt.errorbar(2**(x+4),np.mean(gau_w4,1),np.var(gau_w4,1),lw=2,color='grey', alpha=0.5, label='layer 4')

plt.plot(2**(x+4),0.125*(2**(x+4))**(-0.5),color = 'black',ls='--',lw=2, label =r'$1/\sqrt{n}$')
plt.plot(2**(x+4),0.67*(2**(x+4))**(-0.5),color = 'black',ls='--',lw=2)
plt.plot(2**(x+4),2.80*(2**(x+4))**(-1.0),color = 'red',ls='--',lw=2, label =r'$1/n$')
plt.plot(2**(x+4),2.40*(2**(x+4))**(-1.0),color = 'red',ls='--',lw=2)

ax0.set_xscale("log"); ax0.set_yscale("log")
plt.legend(fontsize=12); plt.xlabel('$n$',fontsize=15); plt.ylabel('$||W_T^l - W_0^l||_F/||W_0^l||_F$',fontsize=15)
plt.title('(a) Gaussian initialization',fontsize=15)

ax1 = plt.subplot((gs[0,1]))

plt.errorbar(2**(x+4),np.mean(ort_w1,1),np.var(ort_w1,1),lw=2,alpha=0.5,color='blue')
plt.errorbar(2**(x+4),np.mean(ort_w2,1),np.var(ort_w2,1),lw=2,alpha=0.5,color='brown')
plt.errorbar(2**(x+4),np.mean(ort_w3,1),np.var(ort_w3,1),lw=2,alpha=0.5,color='green')
plt.errorbar(2**(x+4),np.mean(ort_w4,1),np.var(ort_w4,1),lw=2,alpha=0.5,color='grey')

plt.plot(2**(x+4),0.125*(2**(x+4))**(-0.5),color = 'black',ls='--',lw=2)
plt.plot(2**(x+4),0.665*(2**(x+4))**(-0.5),color = 'black',ls='--',lw=2)
plt.plot(2**(x+4),2.80*(2**(x+4))**(-1.0),color = 'red',ls='--',lw=2)
plt.plot(2**(x+4),2.40*(2**(x+4))**(-1.0),color = 'red',ls='--',lw=2)

plt.xlabel('$n$',fontsize=15); ax1.set_xscale("log"); ax1.set_yscale("log")
plt.title('(b) Orthogonal initialization',fontsize=15)

ax2 = plt.subplot((gs[0,2]))

plt.errorbar(2**(x+4),np.mean(gau_ntk,1),np.var(gau_ntk,1),lw=2,color='blue',label ='Gaussian')
plt.errorbar(2**(x+4),np.mean(ort_ntk,1),np.var(ort_ntk,1),lw=2,color='green',label ='Orthogonal')

plt.plot(2**(x+4),3.5*(2**(x+4))**(-0.5),color = 'black',ls='--',lw=2, label =r'$1/\sqrt{n}$')
plt.plot(2**(x+4),12.0*(2**(x+4))**(-1.0),color = 'red',ls='--',lw=2, label =r'$1/n$')

ax2.set_xscale("log")
ax2.set_yscale("log")
plt.legend(fontsize=12);
plt.xlabel('$n$',fontsize=15); plt.ylabel('$||\hat \Theta_T^{(n)} -\hat \Theta_0^{(n)}||_F/||\hat \Theta_0^{(n)}||_F$',fontsize=15)
plt.title('(c) NTK change',fontsize=15)

plt.savefig("fig2_ntkchange.pdf", bbox_inches='tight')
```
