# Figures 4

Generate figures of dynamics of stochastic gradient descent on Gaussian and orthogonal initialized networks.

## Setup

```python
from __future__ import print_function
import os, sys
from tqdm import tqdm
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
```

## $N = 800$ and $L=5$ using SGD on CIFAR10

```python
import csv   

number_av = 30; number_st = 100000
Ort_N800_train_acc = np.zeros((number_av,number_st)); Ort_N800_train_los = np.zeros((number_av,number_st))
Ort_N800_test_acc = np.zeros((number_av,number_st)); Ort_N800_test_los = np.zeros((number_av,number_st))

for i in range(number_av):
    j = 0

    with open('../data/sgd/tanh_ort_L5_N800/tanh_orthogonal_L5_N800_sgd_%d.csv' %i , newline='') as csvfile:
        spamreader = csv.reader(csvfile, delimiter=' ')
        for row in spamreader:
            
            Ort_N800_train_acc[i,j] = row[2]; Ort_N800_train_los[i,j] = row[3]
            Ort_N800_test_acc[i,j] = row[4];  Ort_N800_test_los[i,j] = row[5] 
            j = j+1
    
Ort_N800_train_acc_av = np.mean(Ort_N800_train_acc[:,:],0)
Ort_N800_test_acc_av = np.mean(Ort_N800_test_acc[:,:],0)

Ort_N800_train_los_av = np.mean(Ort_N800_train_los[:,:],0)
Ort_N800_test_los_av = np.mean(Ort_N800_test_los[:,:],0)
```

```python
import csv   

number_av = 30; number_st = 100000
Gau_N800_train_acc = np.zeros((number_av,number_st)); Gau_N800_train_los = np.zeros((number_av,number_st))
Gau_N800_test_acc = np.zeros((number_av,number_st)); Gau_N800_test_los = np.zeros((number_av,number_st))

for i in range(number_av):
    j = 0

    with open('../data/sgd/tanh_gau_L5_N800/tanh_gaussian_L5_N800_sgd_%d.csv' %i , newline='') as csvfile:
        spamreader = csv.reader(csvfile, delimiter=' ')
        for row in spamreader:
            
            Gau_N800_train_acc[i,j] = row[2]; Gau_N800_train_los[i,j] = row[3]
            Gau_N800_test_acc[i,j] = row[4];  Gau_N800_test_los[i,j] = row[5] 
            j = j+1
    
Gau_N800_train_acc_av = np.mean(Gau_N800_train_acc[:,:],0)
Gau_N800_test_acc_av = np.mean(Gau_N800_test_acc[:,:],0)

Gau_N800_train_los_av = np.mean(Gau_N800_train_los[:,:],0)
Gau_N800_test_los_av = np.mean(Gau_N800_test_los[:,:],0)
```

```python
plt.figure(figsize=(9, 3))
gs = plt.GridSpec(1, 2)
matplotlib.rc('xtick',labelsize=15)
matplotlib.rc('ytick',labelsize=15)
import seaborn as sns
sns.set_style("darkgrid", {"axes.facecolor":".95"})

a1 = plt.subplot(gs[0, 0])

plt.semilogx(Ort_N800_train_acc_av,color='blue',ls='--',label ='Train Orthogonal')
plt.semilogx(Ort_N800_test_acc_av,color='red',ls='--',label ='Test Orthogonal')
plt.semilogx(Gau_N800_train_acc_av,color='blue',label ='Train Orthogonal')
plt.semilogx(Gau_N800_test_acc_av,color='red',label ='Test Orthogonal')

a2 = plt.subplot(gs[0, 1])
plt.plot(Ort_N800_train_los_av,color='blue',ls='--',label = 'Train Gaussian')
plt.plot(Ort_N800_test_los_av,color='red',ls='--',label ='Test Gaussian')
plt.plot(Gau_N800_train_los_av,color='blue',label = 'Train Gaussian')
plt.plot(Gau_N800_test_los_av,color='red',label ='Test Gaussian')
```

## $N = 1600$ and $L=9$ using RMSProp on MNIST

```python
import csv   

number_av = 30; number_st = 12000
Gau_RMSP_train_acc = np.zeros((number_av,number_st)); Gau_RMSP_train_los = np.zeros((number_av,number_st))
Gau_RMSP_test_acc = np.zeros((number_av,number_st)); Gau_RMSP_test_los = np.zeros((number_av,number_st))

for i in range(number_av):
    j = 0

    with open('../data/sgd/relu2_gau_L9_mom_N1600/relu2_gaussian_L9_N1600_mom_%d.csv' %i , newline='') as csvfile:
        spamreader = csv.reader(csvfile, delimiter=' ')
        for row in spamreader:
            
            Gau_RMSP_train_acc[i,j] = row[2]; Gau_RMSP_train_los[i,j] = row[3]
            Gau_RMSP_test_acc[i,j] = row[4];  Gau_RMSP_test_los[i,j] = row[5] 
            j = j+1
    
Gau_RMSP_train_acc_av = np.mean(Gau_RMSP_train_acc[:,:],0)
Gau_RMSP_test_acc_av = np.mean(Gau_RMSP_test_acc[:,:],0)

Gau_RMSP_train_los_av = np.mean(Gau_RMSP_train_los[:,:],0)
Gau_RMSP_test_los_av = np.mean(Gau_RMSP_test_los[:,:],0)
```

```python
import csv   

number_av = 30; number_st = 12000
Ort_RMSP_train_acc = np.zeros((number_av,number_st)); Ort_RMSP_train_los = np.zeros((number_av,number_st))
Ort_RMSP_test_acc = np.zeros((number_av,number_st)); Ort_RMSP_test_los = np.zeros((number_av,number_st))

for i in range(number_av):
    j = 0

    with open('../data/sgd/relu2_ort_L9_mom_N1600/relu2_orthogonal_L9_N1600_mom_%d.csv' %i , newline='') as csvfile:
        spamreader = csv.reader(csvfile, delimiter=' ')
        for row in spamreader:
            
            Ort_RMSP_train_acc[i,j] = row[2]; Ort_RMSP_train_los[i,j] = row[3]
            Ort_RMSP_test_acc[i,j] = row[4];  Ort_RMSP_test_los[i,j] = row[5] 
            j = j+1
    
Ort_RMSP_train_acc_av = np.mean(Ort_RMSP_train_acc[:,:],0)
Ort_RMSP_test_acc_av = np.mean(Ort_RMSP_test_acc[:,:],0)

Ort_RMSP_train_los_av = np.mean(Ort_RMSP_train_los[:,:],0)
Ort_RMSP_test_los_av = np.mean(Ort_RMSP_test_los[:,:],0)
```

```python
plt.figure(figsize=(9, 3))
gs = plt.GridSpec(1, 2)
matplotlib.rc('xtick',labelsize=15)
matplotlib.rc('ytick',labelsize=15)
import seaborn as sns
sns.set_style("darkgrid", {"axes.facecolor":".95"})

a1 = plt.subplot(gs[0, 0])

plt.semilogx(Ort_RMSP_train_los_av[::20],color='blue',ls='--',label = 'Train Gaussian')
plt.semilogx(Ort_RMSP_test_los_av[::20],color='red',ls='--',label ='Test Gaussian')
plt.semilogx(Gau_RMSP_train_los_av[::20],color='blue',label = 'Train Gaussian')
plt.semilogx(Gau_RMSP_test_los_av[::20],color='red',label ='Test Gaussian')


a2 = plt.subplot(gs[0, 1])
plt.semilogx(Ort_RMSP_train_acc_av[::20],color='blue',ls='--',label ='Train Orthogonal')
plt.semilogx(Ort_RMSP_test_acc_av[::20],color='red',ls='--',label ='Test Orthogonal')
plt.semilogx(Gau_RMSP_train_acc_av[::20],color='blue',label ='Train Orthogonal')
plt.semilogx(Gau_RMSP_test_acc_av[::20],color='red',label ='Test Orthogonal')
```

##  $N = 800$ and $L=7$ Orthogonal network 

# Figure 4:  Dynamics of full batch gradient descent on dataset
Orthogonally initialized networks behave similarly to the networks with Gaussian initialization in the NTK regime. (a)(b) We adopt the network architecture of depth of $L=5$, width of $n = 800$, activation of tanh function, with $\sigma^2_w = 2.0$, and $\sigma^2_b =0.1$. The networks are trained by SGD with a small learning rate of $\eta = 10^{-3}$ for $T=10^5$ steps with a batch size of $10^3$ on cross-entropy loss on full CIFAR-10. (c)(d)  We adopt the network architecture of depth of $L=9$, width of $n = 1600$, activation of ReLU function, with $\sigma^2_w = 2.0$, and $\sigma^2_b =0.1$. The networks are trained by PMSProp with a small learning rate of $\eta = 10^{-5}$ for $T=1.2 \times 10^4$ steps with a batch size of $10^3$ on MSE loss on MNIST. While the solid lines stand for Gaussian weights, dotted lines represent orthogonal initialization.

```python

plt.figure(figsize=(15, 3))
gs = plt.GridSpec(1, 4)
matplotlib.rc('xtick',labelsize=15)
matplotlib.rc('ytick',labelsize=15)
import seaborn as sns
sns.set_style("darkgrid", {"axes.facecolor":".95"})

a1 = plt.subplot(gs[0, 0])


plt.semilogx(Gau_N800_train_acc_av,color='blue',label ='Train Gaussian')
plt.semilogx(Gau_N800_test_acc_av,color='red',label ='Test Gaussian')
plt.semilogx(Ort_N800_train_acc_av,color='blue',ls='--',label ='Train Orthogonal')
plt.semilogx(Ort_N800_test_acc_av,color='red',ls='--',label ='Test Orthogonal')

plt.xlim(1, 2e5)
plt.ylim(0.02, 0.8)
plt.legend(fontsize=10); 
plt.xlabel('$t$',fontsize=15); 
plt.title('(a) CIFAR10, Accuracy',fontsize=15)


a2 = plt.subplot(gs[0, 1])


plt.semilogx(Gau_N800_train_los_av,color='blue',label = 'Train Gaussian')
plt.semilogx(Gau_N800_test_los_av,color='red',label ='Test Gaussian')
plt.semilogx(Ort_N800_train_los_av,color='blue',ls='--',label = 'Train Gaussian')
plt.semilogx(Ort_N800_test_los_av,color='red',ls='--',label ='Test Gaussian')

plt.xlim(1, 2e5)
plt.xlabel('$t$',fontsize=15); 
plt.title('(b) CIFAR10, Loss',fontsize=15)

a3 = plt.subplot(gs[0, 2])


plt.semilogx(Gau_RMSP_train_acc_av,color='blue',label ='Train Orthogonal')
plt.semilogx(Gau_RMSP_test_acc_av,color='red',label ='Test Orthogonal')
plt.semilogx(Ort_RMSP_train_acc_av,color='blue',ls='--',label ='Train Orthogonal')
plt.semilogx(Ort_RMSP_test_acc_av,color='red',ls='--',label ='Test Orthogonal')

plt.xlim(1, 2e4)
plt.xlabel('$t$',fontsize=15); 
plt.title('(c) MNIST, Accuracy',fontsize=15)

a4 = plt.subplot(gs[0, 3])


plt.semilogx(Gau_RMSP_train_los_av,color='blue',label = 'Train Gaussian')
plt.semilogx(Gau_RMSP_test_los_av,color='red',label ='Test Gaussian')
plt.semilogx(Ort_RMSP_train_los_av,color='blue',ls='--',label = 'Train Gaussian')
plt.semilogx(Ort_RMSP_test_los_av,color='red',ls='--',label ='Test Gaussian')

plt.xlim(1, 2e4)
plt.xlabel('$t$',fontsize=15); 
plt.title('(d) MNIST, Loss',fontsize=15)

plt.savefig("fig4_sgdnet.pdf", bbox_inches='tight')
```
